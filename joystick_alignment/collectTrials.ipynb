```python
import scipy.io
import pandas as pd
import numpy as np

# Load the .mat file
mat_data = scipy.io.loadmat('FN13_Joystick_V_1_5_1_Opto_20240618_115440 (2).mat')

session_data = mat_data['SessionData']

# Access the 'TrialStartTimestamp' and 'TrialEndTimestamp' fields from SessionData
trial_start_times = np.squeeze(session_data['TrialStartTimestamp'][0, 0])
trial_end_times = np.squeeze(session_data['TrialEndTimestamp'][0, 0])

# Now, loop through and print the start and end times for each trial
for i in range(len(trial_start_times)):
    start_time = trial_start_times[i] - trial_start_times[0]
    end_time = trial_end_times[i] - trial_start_times[0]
    print(f"Trial {i+1} Start: {start_time}, End: {end_time}")

# Create a DataFrame with the start and end times
df = pd.DataFrame({
    'Trial': range(1, len(trial_start_times) + 1),
    'StartTime': trial_start_times - trial_start_times[0],
    'EndTime': trial_end_times - trial_start_times[0]
})

# Save the DataFrame to a CSV file
df.to_csv('trial_times.csv', index=False)

print("Start and end times saved to 'trial_times.csv'.")
```
